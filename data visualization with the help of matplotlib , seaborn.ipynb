```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("uma_task1.xlsx")
```

```python
df.head()
```

```python
plt.figure(figsize =(12,10))
labels =['male','female']
sizes = df.Gender.value_counts().values
explode = (0,0.1)
colors = ['green', 'red']
plt.pie(sizes , explode, labels, colors = colors, autopct = '%2.2f%%', shadow = True , startangle = 0)
plt.title("Gendar Employee Ratio", size = 12)
plt.show()
```

```python
plt.figure(figsize =(12,10))
labels =['Engineering', 'Sales', 'IT', 'Accounting', 'Finance', 'Marketing',
       'Human Resources']
sizes = df.Department.value_counts().values
explode = (0.1,0,0,0,0,0,0)
colors = ['green', 'red','yellow','blue']
plt.pie(sizes , explode, labels, colors = colors, autopct = '%2.2f%%', shadow = True , startangle = 0)
plt.title("Department Employee Ratio", size = 12)
plt.show()
```
